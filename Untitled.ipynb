```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sampling frequency
Fs = 1000

# Time vector
t = np.arange(0, 1, 1/Fs)

# Generate a sine wave with frequency of 50 Hz and amplitude of 1
x = np.sin(2*np.pi*50*t)

# Compute the FFT of the signal
fft_x = np.fft.fft(x)

# Compute the frequency axis
f = np.linspace(0, Fs, len(fft_x))

# Plot the magnitude of the FFT
plt.plot(f, np.abs(fft_x))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.show()

# Compute the iFFT of the FFT
ifft_x = np.fft.ifft(fft_x)

# Plot the original signal and the reconstructed signal
plt.plot(t, x, label='Original Signal')
plt.plot(t, ifft_x.real, label='Reconstructed Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```

```python
def plot_fourier_transform(f,t):
    #a+ib
    fft = np.fft.fft(f)
    fft_mag = np.abs(fft)
    
    ift = np.fft.ifft(fft)
    ift_real = ift.real
    
    #signal plot
    plt.plot(t,f)
    plt.show()

    #fourier plot
    plt.plot(f,fft_mag)
    plt.show()

    #inverse fourier plot
    plt.plot(t,ift_real)
    plt.show()
    

def generatesinwave(Amp,frequency):
    dt = 0.0001
    t = np.arange(0,1,dt)
    #A sin(wt) = A sin(2pift)
    f = Amp*np.sin(2*np.pi*frequency*t)
    return f
    
f= generatesinwave(1,1)+generatesinwave(2,3)+generatesinwave(3,50)
plot_fourier_transform(f,np.arange(0,1,0.0001))
```
